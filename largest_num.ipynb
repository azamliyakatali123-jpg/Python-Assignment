```python
# Finding the largest number in a list.
def largest_num(LIST: list) -> int:
    
    """
    Args:
        This takes a list as input.
    Returns:
        returns the largest number in a list.
    """
    # Check if input is a list
    if not isinstance(LIST,list):
        raise TypeError("Input must be list.")
    
    # Check if list is empty
    if len(LIST) == 0:
        raise ValueError ("List Cannot be empty.")
    
    # Assume first element is the largest initially
    largest_number = LIST[0]
    
    for num in LIST:
        if num > largest_number:
            largest_number = num
    
    # Return the largest number found
    return largest_number

if __name__ == "__main__":
    list1 = [23, 45, 87, 234, 45, 76, 979, 354]
    try:
        # Call the function
        result = largest_num(list1)
        print(f"{result} is the largest.")
    except (TypeError,ValueError) as e:
        # Handle errors
        print(e)
    except Exception as e:
        # Handle unexpected errors
        print(e)
    
    print("This function works")
```

```python
largest_num([100,200,300,400])
```
