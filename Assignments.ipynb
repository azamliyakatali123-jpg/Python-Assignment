```python
#Reverse String function:

def rev_string(string: str) -> str:
    """
    Args:
        Takes string as a input.
    Returns:
        returns a reversed string.
    """
    if not isinstance (string,str):
        raise TypeError ("Input must be string.")
    if len(string) < 10:
        raise ValueError ("must be atleast 10 char")
    rev = ""
    for char in string:
        rev = char + rev
    return rev

try:
    print(rev_string("Hello python world"))
except (TypeError,ValueError) as e:
    print("Error : ",e)
except Exception as e:
    print("Error",e)
print("Hello World")
```
